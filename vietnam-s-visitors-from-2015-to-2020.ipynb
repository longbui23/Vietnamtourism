```python
world_growth_rate = [25,]

for i in range(1, len(world_visitor)):
    growth_rate = round(((world_visitor[i]-world_visitor[i-1])/world_visitor[i-1])*100, 2)
    world_growth_rate.append(growth_rate)
    
world_growth_rate
```

```python
asean_growth_rate = [15,]

for i in range(1, len(asean_visitor)):
    growth_rate = round(((asean_visitor[i] - asean_visitor[i-1])/asean_visitor[i-1])*100, 2)
    asean_growth_rate.append(growth_rate)
    
asean_growth_rate
```

```python
eu_growth_rate = [20,]

for i in range(1, len(eu_visitor)):
    growth_rate = round(((eu_visitor[i]-eu_visitor[i-1])/eu_visitor[i-1])*100, 2)
    eu_growth_rate.append(growth_rate)
    
eu_growth_rate
```

```python
#int var
barWidth = 0.25

# Set position of bar on X axis
br1 = np.arange(len(world_growth_rate))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]

#plot
plt.bar(br1, world_growth_rate, color = 'r', width = barWidth, label = 'world')
plt.bar(br2, asean_growth_rate, color = 'g', width = barWidth, label = 'ASEAN')
plt.bar(br3, eu_growth_rate, color = 'b', width = barWidth, label = 'EU')

#plot bar
plt.title('Number of visitors in ASEAN countries visiting Vietnam from 2015 to 2019')
plt.xlabel('Years')
plt.ylabel('Visitors growth rate')
plt.legend()
plt.xticks([r + barWidth for r in range(len(world_growth_rate))],
        ['2015', '2016', '2017', '2018', '2019'])
```
